```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Dataset
df = pd.read_csv("country_wise_latest.csv")
df.head()
```

```python
# Exploring the Dataset
df.shape, df.info()
df.describe().T
```

```python
# Data Cleaning
df = df.drop(columns=["Deaths / 100 Recovered"])
df
```

```python
df.isna().sum()
```

```python
# Confirmed Rate Across WHO Regions
plt.figure(figsize=(10,7))
sns.boxplot(x="WHO Region", y=np.log1p(df["Confirmed"]), data=df)
plt.title("Distribution of Confirmed COVID-19 Cases by WHO Region")
plt.ylabel("log(Confirmed + 1)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Most Affected Countries 
top_countries = df.nlargest(10, "Confirmed")

plt.figure(figsize=(12, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_countries, palette='Reds_r')
plt.xlabel("Total Confirmed Cases")
plt.ylabel("Country")
plt.title("Top 10 Most Affected Countries by COVID-19")
plt.show()
```

```python
# Top 10 countries for the Death Rate in each region
top10_deaths_by_region = (
    df.sort_values("Deaths / 100 Cases", ascending=False)
    .groupby("WHO Region")
    .head(10)
)

regions = top10_deaths_by_region["WHO Region"].unique()
num_regions = len(regions)

fig, axes = plt.subplots(num_regions, 1, figsize=(10, num_regions*4))

for i, region in enumerate(regions):
    data = top10_deaths_by_region[top10_deaths_by_region["WHO Region"] == region]
    sns.barplot(
        x="Deaths / 100 Cases",
        y="Country/Region",
        data=data,
        palette="Reds_r",
        ax=axes[i]
    )
    axes[i].set_title(f"Top 10 Countries by Death Rate in {region}")
    axes[i].set_xlabel("Death Rate (%)")
    axes[i].set_ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
# Top 10 countries for the Recovery Rate in each region
top10_recovered_by_region = (
    df.sort_values("Recovered / 100 Cases", ascending=False)
    .groupby("WHO Region")
    .head(10)
)

regions = top10_recovered_by_region["WHO Region"].unique()
num_regions = len(regions)

fig, axes = plt.subplots(num_regions, 1, figsize=(10, num_regions*4))

for i, region in enumerate(regions):
    data = top10_recovered_by_region[top10_recovered_by_region["WHO Region"] == region]
    sns.barplot(
        x="Recovered / 100 Cases",
        y="Country/Region",
        data=data,
        palette="Greens_r",
        ax=axes[i]
    )
    axes[i].set_title(f"Top 10 Countries by Recovery Rate in {region}")
    axes[i].set_xlabel("Recovery Rate (%)")
    axes[i].set_ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
#Top 10 Countries with Fastest Spread in each region
#The average number of new infections in the last week compared to the number of previously existing infections.
top10_spread_by_region = (
    df.sort_values("1 week % increase", ascending=False)
    .groupby("WHO Region")
    .head(10)
)

regions = top10_spread_by_region["WHO Region"].unique()
num_regions = len(regions)

fig, axes = plt.subplots(num_regions, 1, figsize=(10, num_regions*4))

for i, region in enumerate(regions):
    data = top10_spread_by_region[top10_spread_by_region["WHO Region"] == region]
    sns.barplot(
        x="1 week % increase",
        y="Country/Region",
        data=data,
        palette="Reds_r",
        ax=axes[i]
    )
    axes[i].set_title(f"Top 10 Countries by Fastest Spread in {region}")
    axes[i].set_xlabel("1 Week % Increase")
    axes[i].set_ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
# Top 10 countries with the number of infections in the last week
top10_change_by_region = (
    df.sort_values("1 week change", ascending=False)
    .groupby("WHO Region")
    .head(10)
)

top10_change_by_region["log1p_change"] = np.log1p(top10_change_by_region["1 week change"])

regions = top10_change_by_region["WHO Region"].unique()
num_regions = len(regions)

fig, axes = plt.subplots(num_regions, 1, figsize=(10, num_regions*4))

for i, region in enumerate(regions):
    data = top10_change_by_region[top10_change_by_region["WHO Region"] == region]
    sns.barplot(
        x="log1p_change",
        y="Country/Region",
        data=data,
        palette="Oranges_r",
        ax=axes[i]
    )
    axes[i].set_title(f"Top 10 Countries with Weekly Increase (log scale) in {region}")
    axes[i].set_xlabel("log(1 + 1 Week Change)")
    axes[i].set_ylabel("Country")

plt.tight_layout()
plt.show()
```

```python
# Relationship between Death Rate and Recovery Rate
# Countries above right  High deaths + High recoveries
# Countries above left  Low deaths + High recovery
# Countries below  have weak recovery rates
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Deaths / 100 Cases",
    y="Recovered / 100 Cases",
    data=df,
    hue="WHO Region",
    size="Confirmed",
    sizes=(35, 500),
    alpha=0.7,
    palette="tab10"
)

plt.title("Relationship between Death Rate and Recovery Rate")
plt.xlabel("Deaths per 100 Cases (%)")
plt.ylabel("Recovered per 100 Cases (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="WHO Region", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
#Top 10 Countries by Confirmed Cases \ Deaths vs Recovery Rates
top10_confirmed = df.nlargest(10, "Confirmed")

plt.figure(figsize=(8,6))
sns.scatterplot(
    x="Deaths / 100 Cases",
    y="Recovered / 100 Cases",
    data=top10_confirmed,
    hue="Country/Region",
    size="Confirmed",
    sizes=(100, 1000),
    alpha=0.8,
    palette="tab10"
)

plt.title("Top 10 Countries by Confirmed Cases \ Deaths vs Recovery Rate")
plt.xlabel("Deaths per 100 Cases (%)")
plt.ylabel("Recovered per 100 Cases (%)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```
